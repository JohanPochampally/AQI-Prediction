```python
df_copy=df_copy.drop(['City'],axis=1)
df_copy=df_copy.drop(['Date'],axis=1)
```

```python
df_copy = df_copy[df_copy['AQI'].notna()]
```

```python
df_copy.head(30)
```

```python
df_copy.isna().sum()
```

```python
df_copy.size
```

```python
df_copy.shape
```

```python
df_copy['AQI_Bucket'].fillna(df_copy['AQI_Bucket'].mode(), inplace = True)
```

```python
df_copy['Benzene'].fillna(df_copy['Benzene'].median(), inplace = True)
df_copy['PM2.5'].fillna(df_copy['PM2.5'].median(), inplace = True)
df_copy['PM10'].fillna(df_copy['PM10'].median(), inplace = True)
df_copy['NO'].fillna(df_copy['NO'].median(), inplace = True)
df_copy['NO2'].fillna(df_copy['NO2'].median(), inplace = True)
df_copy['NOx'].fillna(df_copy['NOx'].median(), inplace = True)
df_copy['NH3'].fillna(df_copy['NH3'].median(), inplace = True)
df_copy['CO'].fillna(df_copy['CO'].median(), inplace = True)
df_copy['SO2'].fillna(df_copy['SO2'].median(), inplace = True)
df_copy['O3'].fillna(df_copy['O3'].median(), inplace = True)
df_copy['Toluene'].fillna(df_copy['Toluene'].median(), inplace = True)
df_copy['Xylene'].fillna(df_copy['Xylene'].median(), inplace = True)
```

```python
df0 = df_copy.copy()
```

```python
index_names = df0[ df0['AQI'] > 500 ].index
  
# drop these row indexes
# from dataFrame
df0.drop(index_names, inplace = True)
  
df0.size
```

```python
plt.figure(figsize=(10,10))
plt.scatter(x='PM2.5',y='AQI',data=df0)
plt.xlabel('PM2.5')
plt.ylabel('AQI')
```

```python
X = df0.iloc[:,:-2].values
y = df0.iloc[:,-2].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 1)
```

# Multiple Regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
regressor = LinearRegression()
```

```python
regressor.fit(X_train, y_train)
```

```python
y_prediction =  regressor.predict(X_test)
print(y_prediction)
```

```python
# importing r2_score module
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
# predicting the accuracy score
score=r2_score(y_test,y_prediction)
print('r2 socre is',score)
print('mean_sqrd_error is=',mean_squared_error(y_test,y_prediction))
print('root_mean_squared error of is=',np.sqrt(mean_squared_error(y_test,y_prediction)))
print('MAE:', mean_absolute_error(y_test, y_prediction))
```

# Decision Tree Regression

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
dtr = DecisionTreeRegressor(random_state=1)
```

```python
dtr_prediction = dtr.predict(X_test)
```

```python
dtr_prediction
```

```python
dtr_score=r2_score(y_test,dtr_prediction)
print('r2 socre is',dtr_score)
print('mean_sqrd_error is=',mean_squared_error(y_test,dtr_prediction))
print('root_mean_squared error of is=',np.sqrt(mean_squared_error(y_test,dtr_prediction)))
print('MAE:', mean_absolute_error(y_test, dtr_prediction))
```

# Random Forest Regression
